## Sales made in each month

```python
overall_sales_data.head()
```

```python
#sorting month in calender order

month_order = ("January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December")
```

```python
overall_sales_data["month_of_the_year"]= pd.Categorical(overall_sales_data["month_of_the_year"], categories = month_order, ordered = True)
```

```python
overall_sales_data["month_of_the_year"]
```

```python
#groupby month and sales

monthly_sales = overall_sales_data.groupby("month_of_the_year")["sales_amount"].sum()

monthly_sales 
```

```python
monthly_sales.reset_index()
```

## Pizza types that are not doing well on the menu

```python
#grouping the the pizza_type

pizza_performance = overall_sales_data.groupby("pizza_type_id").agg({"quantity": "sum", "sales_amount": "sum"})

pizza_performance
```

```python
#sorting the sales_amount in ascending order

pizza_performance.sort_values(by = "sales_amount")
```

```python
#assuming a treshold of 2%, pizza that contribute less to the total sales

total_sales = overall_sales_data["sales_amount"].sum()

pizza_performance["sales_percentage"] = ((pizza_performance["sales_amount"]) / total_sales)

pizza_performance["sales_percentage"]
```

```python
#underperformer less than 2%

under_performers = pizza_performance[pizza_performance["sales_percentage"] < 0.05]

under_performers
```
